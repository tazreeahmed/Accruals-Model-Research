```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
sub = df[["Accruals", "Cash Revenue Growth", "Gross Property Plant and Equipment"]].copy()

import numpy as np

# 1. Check for inf
print(np.isfinite(sub).all())          
print((~np.isfinite(sub)).sum())      

# 2. Look at scale / variance
print(sub.describe())
```
