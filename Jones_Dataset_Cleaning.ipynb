```python
test = df[df["Total Trade Receivables"].notna()]
test = test[test["Total Receivables"].notna()]

pct_diff = ((test["Total Receivables"] - test["Total Trade Receivables"]) / test["Total Receivables"]) * 100
pct_diff.describe()
```

```python
badpct = (pct_diff > 10).sum()
goodpct = (pct_diff < 10).sum()

print(badpct)
print(goodpct)
```

```python
test = df[(df["Total Trade Receivables"].notna()) & (df["Total Receivables"].notna())]
test["pct_diff"] = ((test["Total Receivables"] - test["Total Trade Receivables"]) /test["Total Receivables"]) * 100
good_gvkeys = test[test["pct_diff"] < 10]["gvkey"].unique()
tr_missing = df[(df["Total Trade Receivables"].isna()) & (df["Total Receivables"].notna())]
safe_rows = tr_missing[ tr_missing["gvkey"].isin(good_gvkeys) ]
len(safe_rows)
```

```python
df.loc[safe_rows.index, "Total Trade Receivables"] = df.loc[safe_rows.index, "Total Receivables"]
df.isna().sum()
```
